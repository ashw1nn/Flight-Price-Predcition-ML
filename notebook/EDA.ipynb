## Feature Engineering process

### Date & Time Extraction

```python
final_df["Date"] = final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"] = final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"] = final_df["Date_of_Journey"].str.split("/").str[2]
final_df.head()
```

```python
final_df["Date"] = final_df["Date"].astype(int)
final_df["Month"] = final_df["Month"].astype(int)
final_df["Year"] = final_df["Year"].astype(int)
```

```python
final_df.drop(labels="Date_of_Journey", axis= 1, inplace=True)
```

```python
final_df.head()
```

```python
final_df["Arrival_Time"].str.split(" ").str[0]
```

```python
final_df["Arrival_Time"] = final_df["Arrival_Time"].str.split(" ").str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df["Arrival_Hour"] = final_df["Arrival_Time"].str.split(":").str[0]
```

```python
final_df["Arrival_Minute"] = final_df["Arrival_Time"].str.split(":").str[1]
```

```python
final_df["Arrival_Hour"] = final_df["Arrival_Hour"].astype(int)
final_df["Arrival_Minute"] = final_df["Arrival_Minute"].astype(int)
```

```python
final_df.drop("Arrival_Time", axis= 1, inplace=True)
```

```python
final_df.head(1)
```

```python
final_df["Dep_Hour"] = final_df["Dep_Time"].str.split(":").str[0]
final_df["Dep_Minute"] = final_df["Dep_Time"].str.split(":").str[1]
final_df["Dep_Hour"] = final_df["Dep_Hour"].astype(int)
final_df["Dep_Minute"] = final_df["Dep_Minute"].astype(int)
final_df.drop("Dep_Time", axis= 1, inplace=True)
```

### Stops extraction

```python
final_df["Total_Stops"].unique()
```

```python
final_df["Total_Stops"] = final_df["Total_Stops"].map({"non-stop": 0,"1 stop": 1,"2 stops": 2,"3 stops": 3,"4 stops": 4,"nan": 1,})
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

### Duration extraction

```python
(final_df["Duration"].str.split(" ").str[1].str.split("h").str[0])
```

```python
final_df["Duration_mins"] = final_df["Duration"].str.split(" ").str[1].str.split("m").str[0]
final_df["Duration_mins"].fillna(0, inplace=True)
final_df["Duration_mins"] = final_df["Duration_mins"].astype(int)
```

```python
final_df["Duration_hr"] = final_df["Duration"].str.split(" ").str[0].str.split("h").str[0]
# final_df[final_df["Duration_hr"] == "5m"]
# final_df.drop(index=2661, axis=0, inplace=True)
final_df["Duration_hr"] = final_df["Duration_hr"].astype(int)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df["Duration"] = final_df["Duration_hr"]*60 + final_df["Duration_mins"]
```

```python
final_df.drop(labels="Duration_hr", axis=1, inplace=True)
final_df.drop(labels="Duration_mins", axis=1, inplace=True)
```

```python
final_df.drop(labels="Route", axis=1, inplace=True)
```

```python
final_df.head()
```

```python
Airline = pd.get_dummies(final_df["Airline"])
Source = pd.get_dummies(final_df["Source"])
Destination = pd.get_dummies(final_df["Destination"])
Additional_Info = pd.get_dummies(final_df["Additional_Info"])
final_df_oheEncoded = pd.concat([final_df, Airline, Source, Destination, Additional_Info], axis=1)
final_df_oheEncoded.drop(labels=["Airline", "Source", "Destination", "Additional_Info"], axis=1, inplace=True)
final_df_oheEncoded
```
